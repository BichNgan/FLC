```python
import numpy as np

# Hàm thành viên cho mật độ giao thông
def MD_T(x):
  if x <= 30:
    return 1
  elif x <= 70:
    return (70 - x) / 40
  else:
    return 0

def MD_TB(x):
  if x <= 30:
    return 0
  elif x <= 70:
    return (x - 30) / 40
  elif x <= 100:
    return 1 - (x - 70) / 30
  else:
    return 0

def MD_C(x):
  if x <= 70:
    return 0
  else:
    return (x - 70) / 30

# Hàm thành viên cho thời lượng đèn xanh
def TXA_N(x):
  if x <= 20:
    return 1
  else:
    return 0

def TXA_V(x):
  if x <= 20:
    return 0
  elif x <= 40:
    return (x - 20) / 20
  elif x <= 60:
    return 1
  else:
    return (60 - x) / 20

def TXA_D(x):
  if x <= 40:
    return 0
  elif x <= 60:
    return (x - 40) / 20
  else:
    return 1

# Quy tắc mờ
def rule(MD_A, MD_B):
  if MD_A == "T" and MD_B == "T":
    return TXA_V(MD_A)
  elif MD_A == "T" and MD_B == "TB":
    return TXA_TB(MD_A)
  elif MD_A == "T" and MD_B == "C":
    return TXA_N(MD_A)
  elif MD_A == "TB" and MD_B == "T":
    return TXA_TB(MD_A)
  elif MD_A == "TB" and MD_B == "TB":
    return TXA_V(MD_A)
  elif MD_A == "TB" and MD_B == "C":
    return TXA_TB(MD_A)
  elif MD_A == "C" and MD_B == "T":
    return TXA_D(MD_A)
  elif MD_A == "C" and MD_B == "TB":
    return TXA_V(MD_A)
  else:
    return TXA_N(MD_A)

# Suy luận mờ
def inference(MD_A, MD_B):
  activation = []
  for MD_B_value in ["T", "TB", "C"]:
    activation.append(min(MD_A(MD_A), MD_B(MD_B_value)) * rule(MD_A, MD_B_value))
  return activation

# Giải mờ
def defuzzification(activation):
  numer = sum([activation[i] * TXA_i.centroid for i in range(len(activation))])
  denom = sum(activation)
  if denom == 0:
    return 0
  else:
    return numer / denom

# Tính toán TXA và TXB
def calculate_TXA_TXB(MD_A, MD_B):
  activation = inference(MD_A, MD_B)
  TXA = defuzzification(activation)
  return TXA

# Ví dụ sử dụng
MD_A = "TB"
MD_B = "C"
TXA = calculate_TXA_TXB(MD_A, MD_B)
print(f"Mật độ giao thông trên Đường A: {MD_A}")
print(f"Mật độ giao thông trên Đường B: {MD_B}")
print(f"Thời lượng đèn xanh cho Đường A: {TXA} giây")
```
